```python
def get_box(js:dict):
    edge=[]
    for i in range(len(js['shapes'])):
        X = []
        Y = []
        for j in js['shapes'][i]['points']:
            Y.append(j[0])
            X.append(j[1])
        edge.append([X,Y])
    # edge: [[X1,Y1],[X2,Y2],...]
    
    raw_mask=[]
    for i in edge:
        rr,cc=skimage.draw.polygon(i[0],i[1])
        raw_mask.append([cc,rr])
    # mask: [[colum1,row1],[colum2,row2],...]   
        
    shape=[]
    for i in edge:
        # 此时i是mask的集合中的一个，i中含有两个集合，x坐标集和y坐标集
        temp=[]# 空列表temp同于存放shape的各项参数
        x_min = min(i[0])
        temp.append(x_min)
        
        y_min = min(i[1])
        temp.append(y_min)
        
        x_max=max(i[0])# 对x坐标集求最大值，得到这个mask的横坐标的最大值
        temp.append(x_max)
        
        y_max = max(i[1])
        temp.append(y_max)
#         center=[(x_min+x_max)/2,(y_min+y_max)/2]
#         temp.append(center)
        shape.append(temp)

    # 此时shape包含若干个mask的shape，每个元素内部存放：x_min,x_max,y_min,y_max,center
    # shape: [[x_min1,x_max1,y_min1,y_max1,center],...]
   
    return shape
```

```python
import utils
from torch.utils.data import Dataset
import PIL.Image as Image
import json
import skimage
class MyDataset(Dataset):
    def __init__(self,root_dir,transform=None):
        self.root_dir = root_dir
        self.labels=os.listdir(self.root_dir)
        for i in self.labels:
            if i[-4:]=='.jpg':
                self.labels.remove(i)
        self.transform=transform
        # getitem内置方法用于获取数据中的图片与标签

    def __getitem__(self, index):
        label_path=os.path.join(self.root_dir,self.labels[index])
        js=json.load(open(label_path))
        image_path=os.path.join(self.root_dir,js['imagePath'])
        shape=get_box(js)
        img=Image.open(image_path)
        label=torch.ones(3,1)
        # 这段代码用于将图片进行预处理
        if self.transform is not None:
            img = self.transform(img)
        shape=torch.Tensor(shape)
        return img,label,shape

    def __len__(self):
        return len(self.img_name)
```

```python
import torch
import torch.nn as nn
import torch.nn.functional as F
import torchvision.models as models
import numpy as np

class ResNet(nn.Module):
    def __init__(self,pretrain=True):
        super(ResNet, self).__init__()
        self.model = models.resnet50(pretrained=pretrain)

        # 定义一个hook函数
        def save_output(module, input, output):
            self.buffer = output
        self.model.layer4.register_forward_hook(save_output)# 在第四层中传入该钩子函数，它将自动将第四层的输出钩取出来作为forward函数的输出

    def forward(self, x):
        self.model(x)
        return self.buffer

class RPN(nn.Module):
    def __init__(self):
        super(RPN, self).__init__()
        self.conv1=nn.Sequential(
            nn.Conv2d(2048, 1024, kernel_size=3, padding=1),
            nn.ReLU()
        )
        # 判断框内是否包含对象的卷积层
        # 通过此次判断，将不包含对象的框筛除，进入下一阶段，即分类器和建议框
        self.selectLayer=nn.Sequential(
            nn.Conv2d(1024, 9, kernel_size=1),
            nn.ReLU()
        )
        # 先验框和建议框的区别：先验框并不保证框内有待检测对象，而建议框会保证框内有待检测对象
        # 生成先验框的卷积层
        self.boxLayer = nn.Sequential(
            nn.Conv2d(1024, 36, kernel_size=1),
            nn.ReLU()
        )

    def forward(self,x):
        x=self.conv1(x)
        score=self.selectLayer(x)#-------------[batch,18,w,h]
        score=F.softmax(score,dim=-3)
        box=self.boxLayer(x)
        box=F.softmax(box,dim=-3)
        box*=torch.Tensor(419)
        return score,box

class FasterRCNN(nn.Module):
    def __init__(self):
        super(FasterRCNN, self).__init__()
        self.backbone=ResNet()
        self.rpn=RPN()

    def forward(self,x):
        x=self.backbone(x)
        _,box=self.rpn(x)
        return box
```

```python
def IoU(box1,box2,batch_size=16):# 输入为[batch＊框数＊４]
    """
    该iou计算函数自动将小于0的iou排除，即所有
    """
    # 对不符合左上右下格式的张量进行转换
    # 检查x1是否大于x2
    for i in range(batch_size):
        target=torch.where(box1[i,:,0]>box1[i,:,2])[0]
        temp=box1[i,target,:]
        box1[i,target,0]=temp[:,2]
        box1[i,target,2]=temp[:,0]
    #检查y1是否大于y2
    
        target=torch.where(box1[i,:,1]>box1[i,:,3])[0]
        temp=box1[i,target,:]
        box1[i,target,1]=temp[:,3]
        box1[i,target,3]=temp[:,1]
    
    if box2[0]>box2[2]:
        temp=box2[0]
        box2[0]=box2[2]
        box2[2]=temp
    
    if box2[1]>box2[3]:
        temp=box2[1]
        box2[1]=box2[3]
        box2[3]=temp
    
    #计算box的面积
    s_box1=torch.abs((box1[:,:,2]-box1[:,:,0])*(box1[:,:,3]-box1[:,:,1]))
    s_box2=torch.abs((box2[2]-box2[0])*(box2[3]-box2[1]))
    
    x1=torch.max(box1[:,:,0],box2[0])
    y1=torch.max(box1[:,:,1],box2[1])
    
    x2=torch.min(box1[:,:,2],box2[2])
    y2=torch.min(box1[:,:,3],box2[3])
    zero=torch.tensor(0)
    
    w,h = torch.max(zero,x2-x1),torch.max(zero,y2-y1) 
    s_intersec=w*h
    #s_intersec=s_intersec.unsqueeze(0).unsqueeze(2)
    IoU=s_intersec/(s_box1+s_box2-s_intersec)
    return IoU
    
```
